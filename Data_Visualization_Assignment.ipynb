```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
#   heatmap.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Create synthetic data with 5 features (columns)
n_samples = 100
X = np.random.randn(n_samples, 5)  # Create random data with normal distribution

# Create correlation between features
X[:, 1] = X[:, 0] * 0.8 + np.random.randn(n_samples) * 0.2  # Feature 2 is correlated with Feature 1
X[:, 2] = X[:, 0] * 0.5 + np.random.randn(n_samples) * 0.3  # Feature 3 is correlated with Feature 1
X[:, 3] = X[:, 4] * 0.7 + np.random.randn(n_samples) * 0.4  # Feature 4 is correlated with Feature 5

# Convert to pandas DataFrame for easy handling
df = pd.DataFrame(X, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5'])

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add labels and title
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(20)

# Generate the dataset for sales data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Create DataFrame
df = pd.DataFrame(data)

# Pivot the data to get sales by month and day
pivot_table = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create a heatmap to visualize the variation in sales across months and days
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='.1f', linewidths=0.5)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Day')
plt.title('Sales Variation Across Months and Days')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd

# Generate the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create meshgrid
x, y = np.meshgrid(x, y)

# Calculate z using the function z = sin(sqrt(x^2 + y^2))
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame (this part is optional)
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Add color bar
fig.colorbar(surf)

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(25)

# Generate the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

# Create DataFrame
df = pd.DataFrame(data)

# Bubble chart
plt.figure(figsize=(10, 6))

# The size of the bubbles is proportional to the population
plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.6, c=df['Population'], cmap='viridis', edgecolors="w", linewidth=2)

# Add titles and labels
plt.title('Bubble Chart: GDP vs Population', fontsize=14)
plt.xlabel('GDP (in billions)', fontsize=12)
plt.ylabel('Population (in millions)', fontsize=12)

# Annotate the countries on the chart
for i, country in enumerate(df['Country']):
    plt.annotate(country, (df['GDP'][i], df['Population'][i]), fontsize=10, ha='center', va='center')

# Show the plot
plt.colorbar(label='Population')  # Color bar representing the population
plt.show()
```

```python
#2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
#   markers based on the 'sizes' and 'colors' columns.
```

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np
import pandas as pd

# Generate random x and y values
np.random.seed(42)
x = np.random.rand(100) * 10  # x values between 0 and 10
y = np.random.rand(100) * 10  # y values between 0 and 10

# Generate random sizes and colors
sizes = np.random.randint(10, 100, 100)  # random sizes for the markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], 100)  # random colors for the markers

# Create a DataFrame to store the data
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a Bokeh figure
p = figure(title="Bokeh Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add scatter plot
p.scatter('x', 'y', size='sizes', color='colors', legend_field='colors', source=df, fill_alpha=0.6)

# Customize the plot appearance
p.legend.title = 'Colors'
p.legend.location = 'top_right'

# Output the plot directly to the notebook
output_notebook()  # Use output_notebook() for Colab

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")

# Add a vertical bar renderer
p.vbar(x=fruits, top=counts, width=0.6, color="skyblue", legend_field="Fruits")

# Customize labels and title
p.xaxis.axis_label = "Fruits"
p.yaxis.axis_label = "Count"
p.legend.title = "Fruit Types"
p.legend.location = "top_left"

# Output the plot to the notebook
output_notebook()  # Use output_notebook() for Colab

# Show the plot
show(p)
```

```python
#5.  Create a Bokeh heatmap using the provided dataset.
#    data_heatmap = np.random.rand(10, 10)
#    x = np.linspace(0, 1, 10)
#    y = np.linspace(0, 1, 10)
#    xx, yy = np.meshgrid(x, y)
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import output_file
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import numpy as np

# Generate the dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for use in Bokeh
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Create a ColumnDataSource to hold the data
source = ColumnDataSource(data={'x': x_flat, 'y': y_flat, 'z': z_flat})

# Create a Bokeh figure
p = figure(title="Heatmap",
           x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1))

# Set up the color mapper (mapping z_flat values to colors)
mapper = linear_cmap(field_name='z', palette=Viridis256, low=min(z_flat), high=max(z_flat))

# Add scatter plot for the heatmap using squares
p.scatter(x='x', y='y', size=10, marker="square",
          color=mapper, source=source, line_color="white", alpha=0.6)

# Add a color bar to represent the values
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Use output_notebook() if you're in a Jupyter notebook, or output_file() for a standalone HTML file
output_notebook()  # Uncomment this if you're using Jupyter notebooks

# If you're not in Jupyter, use the output_file() method to save it as an HTML
# output_file("heatmap.html")

# Show the plot
show(p)
```
