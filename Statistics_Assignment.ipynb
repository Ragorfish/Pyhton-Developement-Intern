````python
# 2. Generate a list of 500 integers containing values between 200 to 300 and store it in the variable `int_list2`.
#   After generating the list, find the following:


#  (i) Compare the given list of visualization for the given data:



 #   1. Frequency & Gaussian distribution

 #   2. Frequency smoothened KDE plot

 #   3. Gaussian distribution & smoothened KDE plot


 # (ii) Write a Python function to calculate the range of a given list of numbers.


 # (iii) Create a program to find the variance and standard deviation of a list of numbers.


 # (iv) Implement a function to compute the interquartile range (IQR) of a list of values.


#  (v) Build a program to calculate the coefficient of variation for a dataset.



#  (vi) Write a Python function to find the mean absolute deviation (MAD) of a list of numbers.


#  (vii) Create a program to calculate the quartile deviation of a list of values.



#  (viii) Implement a function to find the range-based coefficient of dispersion for a dataset.


import random
import math
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

def generate_list():
    int_list = [random.randint(90, 130) for _ in range(100)]
    return int_list

def generate_list2():
    int_list2 = [random.randint(200, 300) for _ in range(500)]
    return int_list2

def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    numbers.sort()
    n = len(numbers)
    mid = n // 2
    if n % 2 == 0:
        return (numbers[mid - 1] + numbers[mid]) / 2
    else:
        return numbers[mid]

def calculate_mode(numbers):
    freq = Counter(numbers)
    mode = max(freq, key=freq.get)
    return mode

def calculate_weighted_mean(values, weights):
    weighted_sum = sum(v * w for v, w in zip(values, weights))
    total_weight = sum(weights)
    return weighted_sum / total_weight

def calculate_geometric_mean(numbers):
    product = math.prod(numbers)
    return product ** (1 / len(numbers))

def calculate_harmonic_mean(numbers):
    return len(numbers) / sum(1 / n for n in numbers)

def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

def calculate_trimmed_mean(numbers, percentage):
    numbers.sort()
    trim_count = int(len(numbers) * percentage / 100)
    trimmed_numbers = numbers[trim_count: -trim_count]
    return calculate_mean(trimmed_numbers)

def calculate_range(numbers):
    return max(numbers) - min(numbers)

def calculate_variance(numbers):
    mean = calculate_mean(numbers)
    return sum((x - mean) ** 2 for x in numbers) / len(numbers)

def calculate_std_dev(numbers):
    return math.sqrt(calculate_variance(numbers))

def calculate_iqr(numbers):
    numbers.sort()
    q1 = numbers[len(numbers) // 4]
    q3 = numbers[3 * len(numbers) // 4]
    return q3 - q1

def calculate_coefficient_of_variation(numbers):
    mean = calculate_mean(numbers)
    std_dev = calculate_std_dev(numbers)
    return (std_dev / mean) * 100

def calculate_mad(numbers):
    mean = calculate_mean(numbers)
    return sum(abs(x - mean) for x in numbers) / len(numbers)

def calculate_quartile_deviation(numbers):
    iqr = calculate_iqr(numbers)
    return iqr / 2

def calculate_range_coefficient_of_dispersion(numbers):
    range_val = calculate_range(numbers)
    mean = calculate_mean(numbers)
    return (range_val / mean) * 100

def visualize_data(numbers):
    # Frequency & Gaussian distribution
    plt.figure(figsize=(12, 6))
    sns.histplot(numbers, bins=20, kde=False, color='blue', label='Frequency')
    sns.kdeplot(numbers, color='red', label='Gaussian Distribution')
    plt.title("Frequency & Gaussian Distribution")
    plt.legend()
    plt.show()

    # Frequency smoothened KDE plot
    plt.figure(figsize=(12, 6))
    sns.histplot(numbers, bins=20, kde=True, color='green', label='KDE Plot')
    plt.title("Frequency Smoothened KDE Plot")
    plt.legend()
    plt.show()

    # Gaussian distribution & smoothened KDE plot
    plt.figure(figsize=(12, 6))
    sns.kdeplot(numbers, color='red', label='Gaussian Distribution')
    sns.kdeplot(numbers, color='green', label='Smoothened KDE Plot')
    plt.title("Gaussian Distribution & Smoothened KDE Plot")
    plt.legend()
    plt.show()

# Generate list of integers
int_list = generate_list()
int_list2 = generate_list2()

# Outputs
print("List:", int_list)
print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", calculate_mode(int_list))

# Example weights for weighted mean calculation
weights = [random.randint(1, 10) for _ in range(100)]
print("Weighted Mean:", calculate_weighted_mean(int_list, weights))

# Ensure all values in the list are positive for geometric mean calculation
positive_list = [max(1, n) for n in int_list]
print("Geometric Mean:", calculate_geometric_mean(positive_list))

print("Harmonic Mean:", calculate_harmonic_mean(int_list))
print("Midrange:", calculate_midrange(int_list))
print("Trimmed Mean (10%):", calculate_trimmed_mean(int_list, 10))

# New outputs for int_list2
print("List2 Range:", calculate_range(int_list2))
print("List2 Variance:", calculate_variance(int_list2))
print("List2 Standard Deviation:", calculate_std_dev(int_list2))
print("List2 IQR:", calculate_iqr(int_list2))
print("List2 Coefficient of Variation:", calculate_coefficient_of_variation(int_list2))
print("List2 MAD:", calculate_mad(int_list2))
print("List2 Quartile Deviation:", calculate_quartile_deviation(int_list2))
print("List2 Range Coefficient of Dispersion:", calculate_range_coefficient_of_dispersion(int_list2))

# Visualization
visualize_data(int_list2)
````

````python
#3. Write a Python class representing a discrete random variable with methods to calculate its expected
#   value and variance.

class DiscreteRandomVariable:
    def __init__(self, outcomes):
        """
        Initialize the discrete random variable.

        :param outcomes: A dictionary where keys are outcomes and values are their probabilities.
        :type outcomes: dict
        """
        if not isinstance(outcomes, dict):
            raise ValueError("Outcomes must be a dictionary.")

        if not all(isinstance(prob, (int, float)) and 0 <= prob <= 1 for prob in outcomes.values()):
            raise ValueError("Probabilities must be numbers between 0 and 1.")

        if not abs(sum(outcomes.values()) - 1) < 1e-6:
            raise ValueError("The probabilities must sum to 1.")

        self.outcomes = outcomes

    def expected_value(self):
        """
        Calculate the expected value of the random variable.

        :return: The expected value (mean).
        :rtype: float
        """
        return sum(outcome * prob for outcome, prob in self.outcomes.items())

    def variance(self):
        """
        Calculate the variance of the random variable.

        :return: The variance.
        :rtype: float
        """
        mean = self.expected_value()
        return sum(prob * (outcome - mean) ** 2 for outcome, prob in self.outcomes.items())

# Example usage
if __name__ == "__main__":
    outcomes = {1: 0.2, 2: 0.3, 3: 0.5}
    random_variable = DiscreteRandomVariable(outcomes)

    print("Expected Value:", random_variable.expected_value())
    print("Variance:", random_variable.variance())
````

````python
import random

def roll_die():
    return random.randint(1, 6)

def simulate_die_rolls(num_rolls):
    outcomes = []

    for _ in range(num_rolls):
        outcomes.append(roll_die())

    return outcomes

def calculate_expected_value(outcomes):
    return sum(outcomes) / len(outcomes)

def calculate_variance(outcomes, expected_value):
    squared_diffs = [(outcome - expected_value) ** 2 for outcome in outcomes]
    return sum(squared_diffs) / len(outcomes)

def main():
    num_rolls = int(input("Enter the number of rolls to simulate: "))

    outcomes = simulate_die_rolls(num_rolls)
    expected_value = calculate_expected_value(outcomes)
    variance = calculate_variance(outcomes, expected_value)

    print(f"Simulated outcomes: {outcomes[:min(10, len(outcomes))]}{'...' if len(outcomes) > 10 else ''}")
    print(f"Expected Value: {expected_value:.2f}")
    print(f"Variance: {variance:.2f}")

if __name__ == "__main__":
    main()
````

````python
#5. Create a Python function to generate random samples from a given probability distribution (e.g.,
#   binomial, Poisson) and calculate their mean and variance.

import numpy as np

def generate_samples_and_stats(distribution, params, sample_size=1000):
    """
    Generate random samples from a given probability distribution and calculate their mean and variance.

    Parameters:
        distribution (str): The name of the probability distribution (e.g., 'binomial', 'poisson').
        params (dict): Parameters of the distribution. Example:
            - For 'binomial': {'n': 10, 'p': 0.5}
            - For 'poisson': {'lam': 3}
        sample_size (int): The number of samples to generate.

    Returns:
        dict: A dictionary containing the samples, mean, and variance.
    """
    if distribution == 'binomial':
        if 'n' not in params or 'p' not in params:
            raise ValueError("Binomial distribution requires 'n' and 'p' parameters.")
        samples = np.random.binomial(n=params['n'], p=params['p'], size=sample_size)
    elif distribution == 'poisson':
        if 'lam' not in params:
            raise ValueError("Poisson distribution requires 'lam' parameter.")
        samples = np.random.poisson(lam=params['lam'], size=sample_size)
    else:
        raise ValueError(f"Unsupported distribution: {distribution}")

    mean = np.mean(samples)
    variance = np.var(samples)

    return {
        "samples": samples,
        "mean": mean,
        "variance": variance
    }

# Example usage:
if __name__ == "__main__":
    # Generate samples for a binomial distribution
    binomial_stats = generate_samples_and_stats(
        distribution='binomial',
        params={'n': 10, 'p': 0.5},
        sample_size=1000
    )
    print("Binomial Distribution:")
    print(f"Mean: {binomial_stats['mean']}")
    print(f"Variance: {binomial_stats['variance']}")

    # Generate samples for a Poisson distribution
    poisson_stats = generate_samples_and_stats(
        distribution='poisson',
        params={'lam': 3},
        sample_size=1000
    )
    print("\nPoisson Distribution:")
    print(f"Mean: {poisson_stats['mean']}")
    print(f"Variance: {poisson_stats['variance']}")
````

````python
#6. Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute
#   the mean, variance, and standard deviation of the samples.

import numpy as np

def gaussian_statistics(mean, std_dev, sample_size=1000):
    """
    Generate random numbers from a Gaussian (normal) distribution and compute
    the mean, variance, and standard deviation of the samples.

    Parameters:
        mean (float): The mean of the Gaussian distribution.
        std_dev (float): The standard deviation of the Gaussian distribution.
        sample_size (int): The number of samples to generate (default is 1000).

    Returns:
        dict: A dictionary containing the generated samples, mean, variance, and standard deviation.
    """
    # Generate random samples from a Gaussian distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

    # Calculate statistics
    calculated_mean = np.mean(samples)
    calculated_variance = np.var(samples)
    calculated_std_dev = np.std(samples)

    return {
        "samples": samples,
        "mean": calculated_mean,
        "variance": calculated_variance,
        "std_dev": calculated_std_dev
    }

# Example usage
if __name__ == "__main__":
    # Parameters for the Gaussian distribution
    mean = 5  # Mean of the distribution
    std_dev = 2  # Standard deviation of the distribution
    sample_size = 1000  # Number of samples to generate

    # Generate samples and compute statistics
    stats = gaussian_statistics(mean=mean, std_dev=std_dev, sample_size=sample_size)

    # Print results
    print("Gaussian Distribution:")
    print(f"Generated Mean: {stats['mean']:.4f}")
    print(f"Generated Variance: {stats['variance']:.4f}")
    print(f"Generated Standard Deviation: {stats['std_dev']:.4f}")
````

````python
#8. Write a Python function to calculate the probability density function (PDF) of a continuous random
#  variable for a given normal distribution.

import math

def normal_pdf(x, mu, sigma):
    """
    Calculate the PDF of a normal distribution for a given value of x.

    Parameters:
    - x: The value at which to evaluate the PDF.
    - mu: The mean of the normal distribution.
    - sigma: The standard deviation of the normal distribution.

    Returns:
    - The PDF value at x.
    """
    # Calculate the PDF using the normal distribution formula
    exponent = -((x - mu) ** 2) / (2 * sigma ** 2)
    coefficient = 1 / (math.sqrt(2 * math.pi) * sigma)

    return coefficient * math.exp(exponent)

# Example usage
x_value = 0
mean = 0
std_dev = 1

pdf_value = normal_pdf(x_value, mean, std_dev)
print(f"PDF at x = {x_value}: {pdf_value}")
````

````python
#9. Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

import math

def exponential_cdf(x, lambda_param):
    """
    Calculate the CDF of an exponential distribution for a given value of x.

    Parameters:
    - x: The value at which to evaluate the CDF (x >= 0).
    - lambda_param: The rate parameter (1 / mean of the distribution).

    Returns:
    - The CDF value at x.
    """
    if x < 0:
        return 0  # CDF is 0 for negative values of x

    return 1 - math.exp(-x / lambda_param)

# Example usage
x_value = 2
lambda_value = 1.5  # rate parameter (1/mean)

cdf_value = exponential_cdf(x_value, lambda_value)
print(f"CDF at x = {x_value}: {cdf_value}")
````

````python
#10. Write a Python function to calculate the probability mass function (PMF) of Poisson distribution

import math

def poisson_pmf(k, lambda_param):
    """
    Calculate the PMF of a Poisson distribution for a given value of k.

    Parameters:
    - k: The number of occurrences (k >= 0).
    - lambda_param: The rate parameter (average number of occurrences).

    Returns:
    - The PMF value at k.
    """
    if k < 0:
        return 0  # PMF is not defined for negative values of k

    # Use the Poisson formula
    return (lambda_param ** k * math.exp(-lambda_param)) / math.factorial(k)

# Example usage
k_value = 3
lambda_value = 2.5  # rate parameter (average number of occurrences)

pmf_value = poisson_pmf(k_value, lambda_value)
print(f"PMF at k = {k_value}: {pmf_value}")
````

````python
#11. A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors
#who make a purchase). They collect data from the old and new layouts to compare.


#To generate the data use the following command:

#```python

#import numpy as np

# 50 purchases out of 1000 visitors

#old_layout = np.array([1] * 50 + [0] * 950)

# 70 purchases out of 1000 visitors

#new_layout = np.array([1] * 70 + [0] * 930)

#  ```

#Apply z-test to find which layout is successful.

import numpy as np
import scipy.stats as stats

# Data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Sample sizes
n1 = len(old_layout)
n2 = len(new_layout)

# Proportions of successes (purchases)
p1 = np.mean(old_layout)
p2 = np.mean(new_layout)

# Pooled proportion
p = (np.sum(old_layout) + np.sum(new_layout)) / (n1 + n2)

# Standard error of the difference in proportions
se = np.sqrt(p * (1 - p) * (1/n1 + 1/n2))

# Z-statistic
z = (p1 - p2) / se

# p-value for one-tailed test (because we are testing if new layout is better)
p_value = 1 - stats.norm.cdf(z)

# Output
print(f"Z-statistic: {z}")
print(f"P-value: {p_value}")

# Compare with significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The new layout is more successful.")
else:
    print("Fail to reject the null hypothesis: No significant difference in conversion rates.")
````

````python
#12. M A tutoring service claims that its program improves students' exam scores. A sample of students who
#participated in the program was taken, and their scores before and after the program were recorded.


#Use the below code to generate samples of respective arrays of marks:

#```python

#before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])

#after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

#```

#Use z-test to find if the claims made by tutor are true or false.

import numpy as np
import scipy.stats as stats

# Sample data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate the differences (after - before)
differences = after_program - before_program

# Calculate the sample mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Use ddof=1 for sample standard deviation
n = len(differences)

# Calculate the z-statistic
z = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value for a one-tailed test
p_value = 1 - stats.norm.cdf(z)

# Output the results
print(f"Z-statistic: {z}")
print(f"P-value: {p_value}")

# Significance level
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: The tutoring program improves scores.")
else:
    print("Fail to reject the null hypothesis: No significant improvement in scores.")
````

````python
#13. A pharmaceutical company wants to determine if a new drug is effective in reducing blood pressure. They
#conduct a study and record blood pressure measurements before and after administering the drug.


#Use the below code to generate samples of respective arrays of blood pressure:


#```python

#before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])

#after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

#  ```


#Implement z-test to find if the drug really works or not.

import numpy as np
import scipy.stats as stats

# Sample data
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Calculate the differences (after - before)
differences = after_drug - before_drug

# Calculate the sample mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Use ddof=1 for sample standard deviation
n = len(differences)

# Calculate the z-statistic
z = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value for a one-tailed test (since we are testing if the drug reduces blood pressure)
p_value = stats.norm.cdf(z)

# Output the results
print(f"Z-statistic: {z}")
print(f"P-value: {p_value}")

# Significance level
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: The drug effectively reduces blood pressure.")
else:
    print("Fail to reject the null hypothesis: The drug does not significantly reduce blood pressure.")
````

````python
#14. A customer service department claims that their average response time is less than 5 minutes. A sample
#of recent customer interactions was taken, and the response times were recorded.


#Implement the below code to generate the array of response time:

#```python

#response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])

#```

#Implement z-test to find the claims made by customer service department are tru or false

import numpy as np
import scipy.stats as stats

# Sample data
response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])

# Known population mean (claimed by the customer service department)
mu_0 = 5

# Sample mean and standard deviation
sample_mean = np.mean(response_times)
sample_std = np.std(response_times, ddof=1)  # Use ddof=1 for sample standard deviation
n = len(response_times)

# Calculate the z-statistic
z = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Calculate the p-value for a one-tailed test (since we are testing if the mean is less than 5)
p_value = stats.norm.cdf(z)

# Output the results
print(f"Sample Mean: {sample_mean}")
print(f"Z-statistic: {z}")
print(f"P-value: {p_value}")

# Significance level
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: The average response time is less than 5 minutes.")
else:
    print("Fail to reject the null hypothesis: The average response time is not less than 5 minutes.")
````

````python
#15.  A company is testing two different website layouts to see which one leads to higher click-through rates.
#Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
#freedom, and p-value.


#Use the following data:

#```python

#layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]

#layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

import numpy as np
import scipy.stats as stats

# Data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Calculate the sample means
mean_a = np.mean(layout_a_clicks)
mean_b = np.mean(layout_b_clicks)

# Calculate the sample standard deviations
std_a = np.std(layout_a_clicks, ddof=1)  # Sample standard deviation
std_b = np.std(layout_b_clicks, ddof=1)  # Sample standard deviation

# Sample sizes
n_a = len(layout_a_clicks)
n_b = len(layout_b_clicks)

# Calculate the pooled standard deviation
pooled_std = np.sqrt(((n_a - 1) * std_a**2 + (n_b - 1) * std_b**2) / (n_a + n_b - 2))

# Calculate the t-statistic
t_statistic = (mean_a - mean_b) / (pooled_std * np.sqrt(1/n_a + 1/n_b))

# Calculate the degrees of freedom
degrees_of_freedom = n_a + n_b - 2

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=degrees_of_freedom))

# Output the results
print(f"Mean of Layout A: {mean_a}")
print(f"Mean of Layout B: {mean_b}")
print(f"Standard Deviation of Layout A: {std_a}")
print(f"Standard Deviation of Layout B: {std_b}")
print(f"T-statistic: {t_statistic}")
print(f"Degrees of Freedom: {degrees_of_freedom}")
print(f"P-value: {p_value}")

# Decision
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the click-through rates of Layout A and Layout B.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the click-through rates of Layout A and Layout B.")
````

````python
#16. A pharmaceutical company wants to determine if a new drug is more effective than an existing drug in
#reducing cholesterol levels. Create a program to analyze the clinical trial data and calculate the tstatistic and p-value for the treatment effect.


#Use the following data of cholestrol level:

#```python

#existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]

#new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

import numpy as np
import scipy.stats as stats

# Data
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Calculate the sample means
mean_existing = np.mean(existing_drug_levels)
mean_new = np.mean(new_drug_levels)

# Calculate the sample standard deviations
std_existing = np.std(existing_drug_levels, ddof=1)  # Sample standard deviation
std_new = np.std(new_drug_levels, ddof=1)  # Sample standard deviation

# Sample sizes
n_existing = len(existing_drug_levels)
n_new = len(new_drug_levels)

# Calculate the t-statistic
t_statistic = (mean_existing - mean_new) / np.sqrt((std_existing**2 / n_existing) + (std_new**2 / n_new))

# Calculate the degrees of freedom (using Welch's formula)
df = ((std_existing**2 / n_existing + std_new**2 / n_new) ** 2) / \
     (((std_existing**2 / n_existing) ** 2 / (n_existing - 1)) + ((std_new**2 / n_new) ** 2 / (n_new - 1)))

# Calculate the p-value for a one-tailed test (since we are testing if the new drug is more effective)
p_value = stats.t.cdf(t_statistic, df)

# Output the results
print(f"Mean of Existing Drug: {mean_existing}")
print(f"Mean of New Drug: {mean_new}")
print(f"Standard Deviation of Existing Drug: {std_existing}")
print(f"Standard Deviation of New Drug: {std_new}")
print(f"T-statistic: {t_statistic}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value}")

# Decision
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: The new drug is more effective in reducing cholesterol levels.")
else:
    print("Fail to reject the null hypothesis: The new drug is not more effective in reducing cholesterol levels.")
````

````python
#17. U A school district introduces an educational intervention program to improve math scores. Write a Python
#function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to
#determine if the intervention had a significant impact.


#Use the following data of test score:


#  ```python

#  pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]

#  post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

import numpy as np
import scipy.stats as stats

# Data
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Calculate the differences between post- and pre-intervention scores
differences = np.array(post_intervention_scores) - np.array(pre_intervention_scores)

# Calculate the mean and standard deviation of the differences
mean_difference = np.mean(differences)
std_difference = np.std(differences, ddof=1)  # Sample standard deviation

# Sample size
n = len(differences)

# Calculate the t-statistic
t_statistic = mean_difference / (std_difference / np.sqrt(n))

# Calculate the degrees of freedom
df = n - 1

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df))

# Output the results
print(f"Mean Difference: {mean_difference}")
print(f"Standard Deviation of Differences: {std_difference}")
print(f"T-statistic: {t_statistic}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value}")

# Decision
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: The intervention had a significant impact on the test scores.")
else:
    print("Fail to reject the null hypothesis: The intervention did not have a significant impact on the test scores.")
````

````python
#18. An HR department wants to investigate if there's a gender-based salary gap within the company. Develop
#a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically
#significant difference between the average salaries of male and female employees.


#Use the below code to generate synthetic data:


#```python

# Generate synthetic salary data for male and female employees

#np.random.seed(0)  # For reproducibility

#male_salaries = np.random.normal(loc=50000, scale=10000, size=20)

#female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

import numpy as np
from scipy import stats

# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility

male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

# Calculate the t-statistic and p-value using an independent t-test
t_statistic, p_value = stats.ttest_ind(male_salaries, female_salaries)

# Output the results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Determine if the result is statistically significant at the 0.05 significance level
alpha = 0.05
if p_value < alpha:
    print("There is a statistically significant difference in salaries between male and female employees.")
else:
    print("There is no statistically significant difference in salaries between male and female employees.")
````

````python
#19. A manufacturer produces two different versions of a product and wants to compare their quality scores.
#Create a Python function to analyze quality assessment data, calculate the t-statistic, and decide
#whether there's a significant difference in quality between the two versions.


#Use the following data:


#```python

#version1_scores = [85, 88, 82, 89, 87, 84, 90, 88, 85, 86, 91, 83, 87, 84, 89, 86, 84, 88, 85, 86, 89, 90, 87, 88, 85]

#version2_scores = [80, 78, 83, 81, 79, 82, 76, 80, 78, 81, 77, 82, 80, 79, 82, 79, 80, 81, 79, 82, 79, 78, 80, 81, 82]

import numpy as np
from scipy import stats

def compare_quality(version1_scores, version2_scores, alpha=0.05):
    # Calculate the mean and standard deviation of both versions
    mean1 = np.mean(version1_scores)
    mean2 = np.mean(version2_scores)
    std1 = np.std(version1_scores, ddof=1)  # Sample standard deviation
    std2 = np.std(version2_scores, ddof=1)

    # Calculate the t-statistic and degrees of freedom
    t_stat, p_value = stats.ttest_ind(version1_scores, version2_scores)

    # Output the results
    print(f"Version 1 - Mean: {mean1:.2f}, Std: {std1:.2f}")
    print(f"Version 2 - Mean: {mean2:.2f}, Std: {std2:.2f}")
    print(f"T-statistic: {t_stat:.2f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret the result
    if p_value < alpha:
        print("There is a significant difference in quality between the two versions (reject null hypothesis).")
    else:
        print("There is no significant difference in quality between the two versions (fail to reject null hypothesis).")

# Data
version1_scores = [85, 88, 82, 89, 87, 84, 90, 88, 85, 86, 91, 83, 87, 84, 89, 86, 84, 88, 85, 86, 89, 90, 87, 88, 85]
version2_scores = [80, 78, 83, 81, 79, 82, 76, 80, 78, 81, 77, 82, 80, 79, 82, 79, 80, 81, 79, 82, 79, 78, 80, 81, 82]

# Call the function
compare_quality(version1_scores, version2_scores)
````

````python
#21.  A political analyst wants to determine if there is a significant association between age groups and voter
#preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
#them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
#there is a significant association between age groups and voter preferences.


#Use the below code to generate data:

#```python

#np.random.seed(0)

#age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)

#voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate data
np.random.seed(0)

age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a DataFrame for easier analysis
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform the Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("\nThere is a statistically significant association between age groups and voter preferences.")
else:
    print("\nThere is no statistically significant association between age groups and voter preferences.")
````
